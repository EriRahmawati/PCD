```python
# template matching
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# untuk membaca gambar input.
img_rgb = cv2.imread('rambu.jpg')

plt.imshow(img_rgb)
```

```python


# mengubah gambar ke grayscale.
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)

# untuk membaca gambar template.
template = cv2.imread('rambu_p.png',0)

# untuk menyimpan lebar dan tinggi dari template pada variable w dan h.
w, h = template.shape[::-1]

#  untuk mencockkan antara gambar input dan template.
res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)

# tentukan nilai ambang batas.
threshold = 0.5

# Simpan koordinat area yang cocok dalam array numpy.
loc = np.where( res >= threshold)

# Gambar sebuah kotak di sekitar wilayah yang cocok.
for pt in zip(*loc[::-1]):
	cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (255,0,0), 2)

# menampilkan gambar akhir dengan area yang cocok.
image = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)

pixels = np.array(image)
plt.imshow(pixels)
```
